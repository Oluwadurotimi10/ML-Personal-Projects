Using Neural Networks

```python
import keras
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
```

```python
#encoding the ouput to make it easier for the computer to understand
from sklearn.preprocessing import LabelEncoder #OneHotEncoder
LE = LabelEncoder()
y_NU = LE.fit_transform(y)
y_NU = np_utils.to_categorical(y_NU)
```

```python
y_NU
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y_NU, test_size = 0.2, random_state = 0)
```

```python
def Classifier():
    #Create model
    classifier = Sequential()

    classifier.add(Dense(output_dim = 10,init = 'uniform', activation = 'relu', input_dim = 11))
    #classifier.add(Dropout(p = 0.2))

    classifier.add(Dense(output_dim = 8,init = 'uniform', activation = 'relu'))
    #classifier.add(Dropout(p = 0.2))

    classifier.add(Dense(output_dim = 6,init = 'uniform', activation = 'relu'))
    #classifier.add(Dropout(p = 0.2))
    
    classifier.add(Dense(output_dim = 4,init = 'uniform', activation = 'softmax'))
    
    #Compile model
    classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy']  )
    
    return classifier
```

```python
#calling the classifier function
classifier_NU = KerasClassifier(build_fn = Classifier)
```

```python
#splitting the data into train and validation set
X_train_ori = X_train[:6000]
X_train_val = X_train[6000:]

y_train_ori = y_train[:6000]
y_train_val = y_train[6000:]
```

```python
#training the data
train = classifier_NU.fit(X_train_ori, y_train_ori,batch_size = 10, nb_epoch = 25,validation_data = (X_train_val,y_train_val), verbose =1)
```

```python
y_pred6 = classifier_NU.predict(X_test)
```

```python
y_pred6
```

```python
#Converting prediction to label
pred = list()
for i in range (len(y_pred6)):
    pred.append(np.argmax(y_pred6[i]))
#Converting label encoded test label to label
test = list()
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))
    
```

```python
from sklearn.metrics import classification_report
cr = classification_report(test,pred)
print(cr)
```

```python
plt.plot(train.history["acc"])
plt.plot(train.history["val_acc"])
plt.title("Model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Train", "Test"], loc = "upper left")
plt.show()
```

```python
plt.plot(train.history["loss"])
plt.plot(train.history["val_loss"])
plt.title("Model loss")
plt.ylabel("loss")
plt.xlabel("Epoch")
plt.legend(["Train", "Test"], loc = "upper left")
plt.show()
```
